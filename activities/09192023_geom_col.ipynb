## Bonus❗️

The cases dataset only has case counts for NC. How might you confirm this using `geom_bar`?

```r
NC_counts = cases %>%
    group_by(state) %>%
    count()

p2 = ggplot(NC_counts, aes(y = state)) +
    geom_bar()

p2
```
