```r
options(repr.plot.res = 300, repr.plot.height = 3, repr.plot.width = 4)

p2 = ggplot(diamonds, aes(x = price))

p2 = p2 + stat_ecdf(geom = "point", size = 0.05, alpha = 0.5)

p2 = p2 + geom_hline(aes(yintercept = 0.50), color = 'orange')
p2 = p2 + geom_hline(aes(yintercept = 0.25), color = 'blue')
p2 = p2 + geom_hline(aes(yintercept = 0.75), color = 'red')

p2
```
